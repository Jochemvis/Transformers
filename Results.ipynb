```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# NL-EN
df_nl_en_old = pd.read_excel('/content/nl_en_scores_OLD.xlsx')
df_nl_en_new = pd.read_excel('/content/nl_en_scores_NEW.xlsx')

# FI-EN
df_fi_en_old = pd.read_excel('/content/fi_en_scores_OLD.xlsx')
df_fi_en_new = pd.read_excel('/content/fi_en_scores_NEW.xlsx')

# RU-EN
df_ru_en_old = pd.read_excel('/content/ru_en_scores_OLD.xlsx')
df_ru_en_new = pd.read_excel('/content/ru_en_scores_NEW.xlsx')

# ES-EN
df_es_en_old = pd.read_excel('/content/es_en_scores_OLD.xlsx')
df_es_en_new = pd.read_excel('/content/es_en_scores_NEW.xlsx')

# Rename and scale comet
for df in [df_nl_en_old, df_nl_en_new, df_fi_en_old, df_fi_en_new, df_ru_en_old, df_ru_en_new, df_es_en_old, df_es_en_new]:
    df.rename(columns={'COMET Score': 'comet', 'SacreBLEU Score': 'sacreBLEU', 'CHRF Score': 'CHRF'}, inplace=True)
    df['comet'] = df['comet'] * 100
```

```python
import numpy as np

dataframes = {
    "NL-EN Old": df_nl_en_old,
    "NL-EN New": df_nl_en_new,
    "FI-EN Old": df_fi_en_old,
    "FI-EN New": df_fi_en_new,
    "RU-EN Old": df_ru_en_old,
    "RU-EN New": df_ru_en_new,
    "ES-EN Old": df_es_en_old,
    "ES-EN New": df_es_en_new,
}

for name, df in dataframes.items():
    sacrebleu_avg = np.mean(df['sacreBLEU'])
    chrf_avg = np.mean(df['CHRF'])
    comet_avg = np.mean(df['comet'])
    print(f"{name}: SacreBLEU Avg = {sacrebleu_avg:.2f}, CHRF Avg = {chrf_avg:.2f}, COMET Avg = {comet_avg:.2f}")
```

```python
old_avg = [
    (np.mean(df_nl_en_old['sacreBLEU']), np.mean(df_nl_en_old['CHRF']), np.mean(df_nl_en_old['comet'])),
    (np.mean(df_fi_en_old['sacreBLEU']), np.mean(df_fi_en_old['CHRF']), np.mean(df_fi_en_old['comet'])),
    (np.mean(df_ru_en_old['sacreBLEU']), np.mean(df_ru_en_old['CHRF']), np.mean(df_ru_en_old['comet'])),
    (np.mean(df_es_en_old['sacreBLEU']), np.mean(df_es_en_old['CHRF']), np.mean(df_es_en_old['comet']))
]

new_avg = [
    (np.mean(df_nl_en_new['sacreBLEU']), np.mean(df_nl_en_new['CHRF']), np.mean(df_nl_en_new['comet'])),
    (np.mean(df_fi_en_new['sacreBLEU']), np.mean(df_fi_en_new['CHRF']), np.mean(df_fi_en_new['comet'])),
    (np.mean(df_ru_en_new['sacreBLEU']), np.mean(df_ru_en_new['CHRF']), np.mean(df_ru_en_new['comet'])),
    (np.mean(df_es_en_new['sacreBLEU']), np.mean(df_es_en_new['CHRF']), np.mean(df_es_en_new['comet']))
]

x = np.arange(4)
width = 0.2

# Old
plt.bar(x - width, [d[0] for d in old_avg], width, label='SacreBLEU', color='goldenrod', edgecolor='black')
plt.bar(x, [d[1] for d in old_avg], width, label='CHRF', color='wheat', edgecolor='black')
plt.bar(x + width, [d[2] for d in old_avg], width, label='COMET', color='khaki', edgecolor='black')

plt.xlabel('Language')
plt.ylabel('Average Score')
plt.title('Old Testament Scores')
plt.xticks(x, ['NL-EN', 'FI-EN', 'RU-EN', 'ES-EN'])
plt.legend()
plt.show()

# New
plt.bar(x - width, [d[0] for d in new_avg], width, label='SacreBLEU', color='goldenrod', edgecolor='black')
plt.bar(x, [d[1] for d in new_avg], width, label='CHRF', color='wheat', edgecolor='black')
plt.bar(x + width, [d[2] for d in new_avg], width, label='COMET', color='khaki', edgecolor='black')

plt.xlabel('Language')
plt.ylabel('Average Score')
plt.title('New Testament Scores')
plt.xticks(x, ['NL-EN', 'FI-EN', 'RU-EN', 'ES-EN'])
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

old_comet = [d[2] for d in old_avg]
new_comet = [d[2] for d in new_avg]

x = np.arange(4)
width = 0.35

plt.bar(x - width/2, old_comet, width, label='Old Testament', color='goldenrod', edgecolor='black')
plt.bar(x + width/2, new_comet, width, label='New Testament', color='wheat', edgecolor='black')

plt.xlabel('Language')
plt.ylabel('Average COMET Score')
plt.title('COMET Scores')
plt.xticks(x, ['NL-EN', 'FI-EN', 'RU-EN', 'ES-EN'])
plt.legend()
plt.tight_layout()
plt.show()
```
